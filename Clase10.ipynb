```python
ndf = ndf.drop('nan', axis = 1)
```

```python
ndf.columns
```

### FP growth algorithm

```python
from mlxtend.frequent_patterns import fpgrowth, association_rules
```

```python
fre_itms = fpgrowth(ndf,
                    min_support = 0.6,
                    use_colnames = True)
fre_itms
```

```python
as_ruls = association_rules(fre_itms,
                            metric = 'lift',
                            min_threshold = 1.5)
as_ruls
```

### Caso 1: Which are the top 10 frequent patterns with high Confidence ??

```python
sortConf = as_ruls.sort_values('confidence',
                                ascending = False).head(5)
sortConf
```

### Caso 2: Transactions with antecedents (m)

```python
sortConf[sortConf['antecedents'] == {'m'}]
```

### Caso 3 : Transactions without (c,f) antecedents, and (m) consequents 

```python
ante = as_ruls['antecedents'] == frozenset({'f', 'c'})
```

```python
as_ruls[ante]
```

```python
cons = as_ruls['consequents'] == frozenset({'m'})
```

```python
as_ruls[cons]
```

```python
nos = (ante & cons)
```

```python
as_ruls[nos]
```

```python
as_ruls[~nos]  #con el símbolo de ~ se hace a la inversa
```
